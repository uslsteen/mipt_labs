```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd
```

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
```

```python
###############       data init        ################
lab_name = "lab_5_7_4.xlsx"
list_name = "Лист2"

x_name = "d, см"
x_err_name = "\sigma_d, cm"

y_name2 = "Ni_{мягк}"
y_err_name2 = "\sigma_{Ni_{мягк}, стат}"

data = pd.read_excel(lab_name, sheet_name=list_name)

x_data = data[x_name]
y_data = data[y_name2]

x_err = data[x_err_name]
y_err = data[y_err_name2]

###############       data init        ################

plt.plot(x_data, y_data, 'o', label='measurements')

popt, pcov = curve_fit(func, x_data, y_data)

y = func(x_data, *popt)
# print(y)

plt.plot(x_data, func(x_data, *popt), 'r-',
         label='fit: a=%5.2f, b=%5.2f, c=%5.2f' % tuple(popt))

plt.errorbar(x_data, y_data, xerr= x_err, yerr= y_err, fmt = 'o', color = 'blue')

plt.xlabel("d, см")
plt.ylabel("Ni. мягкая")
plt.legend()
plt.show()
```

```python
###############       data init        ################
lab_name = "lab_5_7_4.xlsx"
list_name = "Лист1"

x_name = "d, см"

x_err_name = "\sigma_d, cm"
y_name1 = "Ni"
y_err_name1 = "\sigma_{Ni, стат}"

data = pd.read_excel(lab_name, sheet_name=list_name)

x_data = data[x_name]
y_data = data[y_name1]

x_err = data[x_err_name]
y_err = data[y_err_name1]


###############       data init        ################

plt.plot(x_data, y_data, 'o', label='measurements')

popt, pcov = curve_fit(func, x_data, y_data, maxfev = 10000)
popt

y = func(x_data, *popt)
# print(y)

plt.plot(x_data, y, 'r-',
         label='fit: a=%5.2f, b=%5.2f, c=%5.2f' % tuple(popt))

plt.errorbar(x_data, y_data, xerr= x_err, yerr= y_err, fmt = 'o', color = 'blue')

plt.xlabel("d, см")
plt.ylabel("Ni")
plt.legend()
plt.show()
```
